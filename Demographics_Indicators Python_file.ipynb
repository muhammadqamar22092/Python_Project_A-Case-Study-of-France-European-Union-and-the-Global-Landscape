```python
import pandas as pd
import matplotlib.pyplot as plt
# Load the datasets
urban_population_df = pd.read_excel('urban_population.xlsx', index_col = 'Years')
life_expectancy_df = pd.read_excel('Life_Expectancy.xlsx',index_col = 'Years')
fertility_rate_df = pd.read_excel('Fertility_rate.xlsx', index_col = 'Years')
population_total_df = pd.read_excel('Population_total.xlsx', index_col = 'Years')
population_growth_df =pd.read_excel('Population_growth.xlsx', index_col = 'Years')
```

```python
# Descriptive analysis
print(urban_population_df.describe())
print(life_expectancy_df.describe())
print(fertility_rate_df.describe())
print(population_total_df.describe())
print(population_growth_df.describe())
```

```python
# Plotting the Urban Population data
plt.figure(figsize=(8, 4))
plt.plot(urban_population_df.index, urban_population_df['World'], label='World', color='blue', marker='*')
plt.plot(urban_population_df.index, urban_population_df['France'], label='France', color='red', marker='+')
plt.plot(urban_population_df.index, urban_population_df['European Union'], label='European Union', color='green', marker='>')

plt.title('Urban Population (% of Total Population)')
plt.xlabel('Year')
plt.ylabel('Urban Population')
plt.axvline(x=1971, color='gray', linestyle='--')
plt.axvline(x=2022, color='purple', linestyle='--')  

# Get the urban population values for France, World, and European Union in 1970 and 2021
urban_population_france_1970 = urban_population_df.loc[1970, 'France']
urban_population_world_1970 = urban_population_df.loc[1970, 'World']
urban_population_eu_1970 = urban_population_df.loc[1970, 'European Union']
urban_population_france_2021 = urban_population_df.loc[2021, 'France']
urban_population_world_2021 = urban_population_df.loc[2021, 'World']
urban_population_eu_2021 = urban_population_df.loc[2021, 'European Union']

# Define custom legend entries with specific colors for France, World, and European Union
custom_legend = [
    plt.Line2D([0], [0], color='red', marker='o', label=f'France: {urban_population_france_1970:.2f}% (1970) - {urban_population_france_2021:.2f}% (2021)'),
    plt.Line2D([0], [0], color='blue', marker='o', label=f'World: {urban_population_world_1970:.2f}% (1970) - {urban_population_world_2021:.2f}% (2021)'),
    plt.Line2D([0], [0], color='green', marker='o', label=f'EU: {urban_population_eu_1970:.2f}% (1970) - {urban_population_eu_2021:.2f}% (2021)'),
]

# Combine the custom legend entries with the existing legend
plt.legend(handles=custom_legend)

plt.grid(True)

plt.savefig('urban_pop_plot.png')
# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the additional Excel files
additional_files = {
    "Age_dependency_ratio": "Age_dependency_ratio.xlsx",
    "Air_pollution": "Air_pollution.xlsx",
    "Renewable_energy_consumption": "Renewable_energy_consumption.xlsx"
}

# Read the files into pandas dataframes
additional_dfs = {name: pd.read_excel(path) for name, path in additional_files.items()}

# Function to process each dataframe by dividing it into decades and calculating the mean for each decade
def process_decade_data(df):
    df['Decade'] = (df['Years'] // 10) * 10
    decade_df = df.groupby('Decade').mean()
    decade_df = decade_df.drop(columns=['Years'], errors='ignore')
    return decade_df

# Process each dataframe
additional_processed_dfs = {name: process_decade_data(df) for name, df in additional_dfs.items()}

# Create a bar chart for each processed dataframe
fig, axes = plt.subplots(nrows=len(additional_processed_dfs), ncols=1, figsize=(8, 10))

# Plotting each dataframe in a separate subplot
for ax, (name, df) in zip(axes.flatten(), additional_processed_dfs.items()):
    df.plot(kind='bar', ax=ax, legend=True)
    ax.set_title(name.replace('_', ' ').title())
    ax.set_ylabel('Value')
    ax.set_xlabel('Decade')

# Adjust the layout and save the figure
plt.tight_layout()
plt.show()

# Output the path to the saved chart
print(chart_file_path)
```
