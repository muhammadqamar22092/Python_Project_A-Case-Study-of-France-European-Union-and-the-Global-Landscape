```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the GHG emissions data
co2_emissions_df = pd.read_excel('CO2_emissions.xlsx', index_col='Years')

# Load the population data
population_data = pd.read_excel('population.xlsx', index_col='Years')

# Merging GHG emissions data with population data
combined_data = pd.merge(co2_emissions_df, population_data, on='Years', how='left')

# Computing GHG emissions per capita
combined_data['France_per_capita'] = combined_data['France'] / combined_data['France_pop']
combined_data['World_per_capita'] = combined_data['World'] / combined_data['World_pop']
combined_data['European Union_per_capita'] = combined_data['European Union'] / combined_data['European Union_pop']

# Create a figure
plt.figure(figsize=(8, 4))

# Plot the data for France, World, and European Union
plt.plot(combined_data.index, combined_data['France_per_capita'], label='France', marker='o')
plt.plot(combined_data.index, combined_data['World_per_capita'], label='World', marker='x')
plt.plot(combined_data.index, combined_data['European Union_per_capita'], label='European Union', marker='x')

# Set the title and labels
plt.title('CO2 Emissions Per Capita (France, World, and European Union)')
plt.xlabel('Year')
plt.ylabel('CO2 Emissions Per Capita (Metric Tons)')

# Get the GHG emissions per capita values for 1990 and 2020
france_1990 = combined_data.loc[1990, 'France_per_capita']
world_1990 = combined_data.loc[1990, 'World_per_capita']
eu_1990 = combined_data.loc[1990, 'European Union_per_capita']
france_2020 = combined_data.loc[2020, 'France_per_capita']
world_2020 = combined_data.loc[2020, 'World_per_capita']
eu_2020 = combined_data.loc[2020, 'European Union_per_capita']

# Add custom legend entries for 1990 and 2020 with values formatted to 3 decimal places
plt.legend([
    f'France (1990: {france_1990:.9f}, 2020: {france_2020:.9f})',
    f'World (1990: {world_1990:.9f}, 2020: {world_2020:.9f})',
    f'European Union (1990: {eu_1990:.9f}, 2020: {eu_2020:.9f})'
])

plt.grid(True)

# Save the plot as a PNG file
plt.savefig('CO2_emissions_per_capita_plot.png')

# Show the plot
plt.show()
```
