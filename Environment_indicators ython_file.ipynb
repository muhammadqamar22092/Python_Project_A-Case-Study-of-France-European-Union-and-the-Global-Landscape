```python
import pandas as pd
import matplotlib.pyplot as plt

# Loading the datasets
forest_area_df = pd.read_excel('Forest_on_land.xlsx', index_col = 'Years')
air_pollution_df = pd.read_excel('Air_pollution.xlsx', index_col ='Years')
agricultural_land_df = pd.read_excel('Agriculture_land.xlsx',index_col = 'Years')
cereal_yield_df = pd.read_excel('Cereal_yield.xlsx', index_col = 'Years')
renewable_energy_df = pd.read_excel('Renewable_energy_consumption.xlsx', index_col = 'Years')
co2_emissions_df = pd.read_excel('CO2_emissions.xlsx', index_col = 'Years')
```

```python
# Descriptive analysis
print("forest_area desciptive states:")
print(forest_area_df.describe())
print("air_pollution_df descriptive stats:")
print(air_pollution_df.describe())
print("agriculture_land_descriptive stats:")
print(agricultural_land_df.describe())
print("creal_yield_df descriptive_stats:")
print(cereal_yield_df.describe())
print("CO2 Emissions Descriptive Statistics:")
print(co2_emissions_df.describe())
print("\nRenewable Energy Consumption Descriptive Statistics:")
print(renewable_energy_df.describe())
```

```python
# Checking for missing values
missing_values_forest_area_df= forest_area_df.isnull().sum()
missing_values_air_pollution_df = air_pollution_df.isnull().sum()
missing_values_agricultural_land_df = agricultural_land_df.isnull().sum()
missing_values_cereal_yield_df = cereal_yield_df.isnull().sum()


```

```python
# Treating missing values
# Forest Area Data
forest_area_df['World'] = forest_area_df['World'].fillna(forest_area_df['World'].mean())

# Air Pollution Data
air_pollution_df.interpolate(method='linear', inplace=True)
```

```python
plt.figure(figsize=(8, 4))
plt.plot(air_pollution_df.index, air_pollution_df['World'], label='World', color='blue',marker = "*")
plt.plot(air_pollution_df.index, air_pollution_df['France'], label='France', color='red',marker = ">")
plt.plot(air_pollution_df.index, air_pollution_df['European Union'], label='European Union', color='green', marker = "+")
plt.title('Air Pollution Trends (Mean Annual Exposure, 1990-2020)')
plt.xlabel('Year')
plt.ylabel('Mean Annual Exposure (micrograms per cubic meter)')

# Add custom legend entries for 1990 and 2021
plt.legend([
    f'World (1990: {air_pollution_df.loc[1990, "World"]:.2f}, 2019: {air_pollution_df.loc[2019, "World"]:.2f})',
    f'France (1990: {air_pollution_df.loc[1990, "France"]:.2f}, 2019: {air_pollution_df.loc[2019, "France"]:.2f})',
    f'European Union (1990: {air_pollution_df.loc[1990, "European Union"]:.2f}, 2019: {air_pollution_df.loc[2019, "European Union"]:.2f})'
])

plt.grid(True)

# Save the plot as a PNG file
plt.savefig('air_pollution_plot.png')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(forest_area_df.index, forest_area_df['World'], label='World', color='blue',marker = "*")
plt.plot(forest_area_df.index, forest_area_df['France'], label='France', color='red',marker = "+")
plt.plot(forest_area_df.index, forest_area_df['European Union'], label='European Union', color='green',marker = ">")
plt.title('Forest Area Trends (1990-2021)')
plt.xlabel('Year')
plt.ylabel('Forest Area (% of Land Area)')

# Add custom legend entries for 1990 and 2021
plt.legend([
    f'World (1990: {forest_area_df.loc[1990, "World"]:.2f}, 2021: {forest_area_df.loc[2021, "World"]:.2f})',
    f'France (1990: {forest_area_df.loc[1990, "France"]:.2f}, 2021: {forest_area_df.loc[2021, "France"]:.2f})',
    f'European Union (1990: {forest_area_df.loc[1990, "European Union"]:.2f}, 2021: {forest_area_df.loc[2021, "European Union"]:.2f})'
])

plt.grid(True)

# Save the plot as a PNG file
plt.savefig('forest_area_plot.png')

# Show the plot
plt.show()
```

```python

# Make sure the DataFrame has columns for 'France', 'World', and 'European Union'

# Calculate the sum of FDI inflows for each entity across the years
total_france_forest_area = forest_area_df['France'].sum()
total_world_forest_area = forest_area_df['World'].sum()
total_eu_forest_area = forest_area_df['European Union'].sum()

# Create a bar chart to visualize the sum values
entities = ['France', 'World', 'European Union']
total_forest_area = [total_france_forest_area, total_world_forest_area, total_eu_forest_area]

plt.figure(figsize=(6, 3))
bars = plt.bar(entities, total_forest_area, color=['blue', 'green', 'orange'])
plt.xlabel('Country')
plt.ylabel('Forest area % of land')
plt.title('Total Forest Area)')
plt.xticks(rotation=45)

# Add legends showing the values on top of the bars
for bar, value in zip(bars, total_forest_area):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.1, bar.get_height(), f'{value:,.0f}', fontsize=8)

# Show the chart

plt.show()
```
